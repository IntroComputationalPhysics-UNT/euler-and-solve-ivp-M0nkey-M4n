# Problem 1:
## Rigid Pendulum Problem Part 1

This pseudocode is to help represent the step-by-step process for solving initial value problems/temporal evolution of systems. For this problem, we take two different sets of inital conditions, `theta_0 = math.pi` and `ang_vel_0 = 0`, and `theta_0 = 0` and `ang_vel_0 = 2`.

For the Euler code of this problem:
1. **Import Math Modules:** Import `math` for simple functions and constants, as well as import `matplotlin.pyplot` for plotting the temporal evolution of the systems onto a graph.
2. **Define the Differential Equation as a First-Order System:** Made a custom function from $\ddot{\theta} + \omega_0^2 sin(\theta)=0$ and made a change of variable to the differential equation which is defined as $\dot{\theta} = \Omega$ and $\dot{\Omega} = - \omega_0^2 sin(\theta)$.
3. **Assign Variables and Time Span:** Define `omega_0`, the initial conditions (`theta_0`: initial angle; and `ang_vel_0`: initial angular velocity), and the time span.
4. **Solve Differential Equation Numerically:** Call our custome Euler solver.
5. **Plot Results and Test Output:** Plot the temporal evolution of `theta` (`theta` vs. `t`), and the phase portrait (`ang_vel` vs. `theta`). Define and compare with an analytic solution to the harmonic oscillator/small-angle approximation.

```python
# Step 1: Import Math Modules (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

import math
import matplotlib.pyplot as plt
```

```python
# Step 2: Define the Differential Equation as a First-Order System (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

def euler_rigid_pendulum(omega_0, theta_0, ang_vel_0, t_0, t_f, n_steps):
  """
  This calculates the solution to theta'' = - omega_0^2 sin(theta) x with Euler's method.
  """
  dt = (t_f - t_0) / n_steps # calculate dt
  theta = [theta_0] # initialized theta list
  ang_vel = [ang_vel_0] # initilaized ang_vel list
  t = [t_0] # initialized t list
  for _ in range(n_steps):
    theta_new = theta[-1] + ang_vel[-1] * dt # update theta
    ang_vel_new = ang_vel[-1] - omega_0 ** 2 * math.sin(theta[-1]) * dt # update angular velocity : using an the Simple Euler's Method
    #ang_vel_new = ang_vel[-1] - omega_0 ** 2 * math.sin(theta_new) * dt # update angular velocity : using the Symplectic/Semi-Implicit Euler's Method
    t_new = t[-1] + dt # update t
    theta.append(theta_new) # append theta
    ang_vel.append(ang_vel_new) # append ang_vel
    t.append(t_new) # append t
  return t, theta, ang_vel
```

```python
# Step 3: Assign Variables and Time Span (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# assign variables and call function
omega_0 = 1
theta_0 = math.pi
ang_vel_0 = 0
t_0 = 0
t_f = 100
n_steps = 5000
```

```python
# Step 4: Solve Differential Equation Numerically (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# call Euler solver
t, theta, ang_vel = euler_rigid_pendulum(omega_0, theta_0, ang_vel_0, t_0, t_f, n_steps)

# get the exact values
theta_exact_sho = [theta_0 * math.cos(omega_0 * (t[i]- t_0)) + (ang_vel_0 / omega_0) * math.sin(omega_0 * (t[i]- t_0)) for i in range(len(t))] # using list comprehension
ang_vel_exact_sho = [-theta_0 * omega_0 * math.sin(omega_0 * (t[i]- t_0)) + ang_vel_0 * math.cos(omega_0 * (t[i]- t_0)) for i in range(len(t))] # using list comprehension
```

```python
# Step 5: Plot Results and Test Outputs (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# plot with matplotlib
plt.figure(figsize=(10,2))
plt.plot(t, theta, label = 'Euler -- rigid pendulum')
plt.plot(t, theta_exact_sho, label = 'Exact')
plt.xlim([0, 100]) # time
plt.ylim([- 5 * math.pi - 0.5, 5 * math.pi + 0.5]) # theta
plt.xlabel('t')
plt.ylabel('$\\theta$')
plt.legend()
plt.show()
```

# Problem 2:
## Rigid Pendulum Problem Part 2

Like Part 1, this pseudocode is to help represent the step-by-step process for solving inital value problems/temporal evolution of systems. However, in this part, we take a somewhat different approach with a different an additional package. This part will also use the same set of initial conditions from Part 1 as well.

1. **Import Packages:** Import `numpy` for constants, special functions, and linear algebra; import `matplotlib.pyplot` for plotting; import `scipy`. `integrate` for solving the differential equation (`solve_ivp`)
2. **Defined the Differential Equation:** Create a first-order system of equations. Define a function which takes in time `t`, and dependent variables `y`, and gives `dy/dt`. In our case, `y = [theta, ang_vel]`.
3. **Define the Initital Conditions, Input Paramters, and the Time Span:** Define `omega_0`, `y_0`, and `t_span`. The initial conditions are also a vector/array (`y_0 = [theta_0, ang_vel_0]`).
4. **Solve the Differential Equation:** Call `solve_ivp`.
5. **Create List to Help Plot**: Create lists to help plot the solutions.
6. **Plot:** Plot the temporal evolution and a phase portrait.

```python
# Step 1: Import Packages (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# import packages
import math
import numpy as np # imports the `numpy` package and gives it an alias/name `np`
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp # imports the solve_ivp function from the `integrate` submodule of the `scipy` module
```

```python
# Step #2: Define the Differential Equation (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# define our differential equation
def rigid_pendulum_ode(t, y, omega_0=1):
  """
  Returns the time derivative of y = [theta, ang_vel] (the angle and angular velocity).
  """
  theta, ang_vel = y # unpacks y = [theta, ang_vel]
  dtheta_dt = ang_vel # define first derivative
  d2theta_dt2 = - omega_0 ** 2 * np.sin(theta) # define second derivative
  dy_dt = [dtheta_dt, d2theta_dt2] # derivative of state variable
  return dy_dt
```

```python
# Step #3: Define the Initial Conditions, Input Parameters, and the Time Span (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# Define the initial conditions, input parameters, and the time span
theta_0 = math.pi # initial angle
ang_vel_0 = 0 # initial angular velocity
y_0 = [theta_0, ang_vel_0] #initial state

# input parameters
omega_0 = 1 # natural frequency

# time span
t_initial = 0 # initial time
t_final = 100 # final time
t_span = (t_initial, t_final) # time span
```

```python
# Step #4: Solve the Differential Equation (Case where `theta_0 = math.pi` and `ang_vel_0 = 0`)

# solve differential equation
solution = solve_ivp(rigid_pendulum_ode, t_span, y_0, args=(omega_0,), dense_output = True, max_step = 1)
```

```python
# Step #5: Create List to Help Plot (Case where `theta_0 = math.pi` and `ang_vel = 0`)

# create list for plotting our solution
n_steps = 1000 # number of time steps
t_plot = np.linspace(t_initial, t_final, n_steps) # list of times for plotting
theta_plot = solution.sol(t_plot)[0] # solution for plotting: angle
ang_vel_plot = solution.sol(t_plot)[1] # solution for plotting: angular velocity
```

```python
# Step #6: Plot (Case where `theta_0 = math.pi' and `ang_vel_0 = 0`)

# plot the temporal evolution
plt.figure(figsize=(10,2))
plt.plot(t_plot, theta_plot, label = 'Rigid Pendulum (Angle vs. Time)')
plt.xlabel('time')
plt.ylabel("$\\theta$")
plt.xlim([0, 100]) # time
plt.ylim([- 5 * np.pi - 0.5, 5 * np.pi + 0.5]) # theta
plt.legend()
plt.show()
```
